```python
def normalize_text(s: pd.Series):
    return (
        s.astype("string").str.strip().str.casefold().str.split().str.join(" "))


def apply_mapping(s: pd.Series, mapping: dict):
    return s.map(lambda x: mapping.get(x, x))
```

```python
status_norm = normalize_text(read_data["status"])
mapping = {"paid": "paid","PAID":"paid" ,"refund": "refund", "refunded": "refund","Refund":"refund"}
status_clean = apply_mapping(status_norm, mapping)
```

```python
read_data["status"]= status_clean
```

```python
refunded = read_data[read_data["status"] == "refund"].groupby("country").size().reset_index(name="refund")
total = pd.DataFrame(read_data.groupby("country").size().reset_index(name="total_orders"))
final = total.merge(refunded, on="country", how="left")
```

```python
final
```

```python
final["refund Rate"] = (final["refund"] / final["total_orders"]) * 100
final["refund Rate"] = final["refund Rate"].round(2)
final
```

```python
# Q1 Which country has the highest number of refunded orders?
final["label"] = (
    final["refund"].astype(int).astype(str)
    + " / "
    + final["total_orders"].astype(int).astype(str)
    + " ("
    + final["refund Rate"].round(1).astype(str)
    + "%)"
)

fig1 = px.bar(
    final,
    x="refund Rate",
    y="country",
    orientation="h",
    title="Which country has the highest number of refunded orders? (%)",
    text="label"
)

fig1.update_traces(textposition="outside")
fig1.update_layout(
    xaxis_title="Refund Rate (%)",
    yaxis_title="Country"
)

fig1.update_layout(
    width=1400,
    height=700
)

fig1.show()
```

```python
save_fig(fig1, FIGS / "refund_rate.png", scale=3)
```

```python
# Q4 Which countries contribute the most to total revenue?
rev = (
    read_data.groupby("country", dropna=False)
      .agg(
          n=("order_id","size"),
          revenue=("amount","sum"),
          aov=("amount","mean"),
      )
      .reset_index()
      .sort_values("revenue", ascending=False)
)

fig2 = px.bar(rev, x="country", y="revenue", title="Which countries contribute the most to total revenue?")
fig2.update_layout(title={"x": 0.02})
fig2.update_xaxes(title_text="Country")
fig2.update_yaxes(title_text="Revenue (sum of amount)")
```

```python
def bootstrap_diff_means(a: pd.Series, b: pd.Series, *, n_boot: int = 2000, seed: int = 0) -> dict:
    rng = np.random.default_rng(seed)
    a = pd.to_numeric(a, errors="coerce").dropna().to_numpy()
    b = pd.to_numeric(b, errors="coerce").dropna().to_numpy()
    assert len(a) > 0 and len(b) > 0, "Empty group after cleaning"

    diffs = []
    for _ in range(n_boot):
        sa = rng.choice(a, size=len(a), replace=True)
        sb = rng.choice(b, size=len(b), replace=True)
        diffs.append(sa.mean() - sb.mean())
    diffs = np.array(diffs)

    return {
        "diff_mean": float(a.mean() - b.mean()),
        "ci_low": float(np.quantile(diffs, 0.025)),
        "ci_high": float(np.quantile(diffs, 0.975)),
    }

d = read_data.assign(is_refund=read_data["status_clean"].eq("refund").astype(int))

a = d.loc[d["country"].eq("SA"), "is_refund"]
b = d.loc[d["country"].eq("AE"), "is_refund"]

print("n_SA:", len(a), "n_AE:", len(b))
res = bootstrap_diff_means(a, b, n_boot=2000, seed=0)
print(res)
```
